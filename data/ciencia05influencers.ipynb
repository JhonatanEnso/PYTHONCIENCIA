```python
import pandas as pd
import numpy as np

df = pd.read_csv('data/influencers.csv')
df
```

```python
df.info()
```

```python
df_filtrado = df.dropna()
df_filtrado
```

limpiar un dataframe y poner una valor por defecto a todos los datos NaN

```python
df_filtrado = df.fillna(0)
df_filtrado
```

```python
df[(df['favorites'] > 400) & (df['mentions'] >= 20)]
```

```python
def popularidad(fila):
    followers = fila['followers']
    followees = fila['followees']
    resultado = followees / followers
    return resultado

df['popularidad'] = df.apply(popularidad, axis=1)
df[df['popularidad'] >= 15]
```

vamos agrupar por el campo COUNTRY y vamos a mostrar la media de dicho grupo 

```python
grupo = df.groupby(['country']).agg({
    "followers": "sum",
    "mentions": "mean",
    "retweets": "max",
})

grupo
```

```python
grupo[grupo['followers'] > 5000]
```

vamos a almacenar en csv


```python
grupo.to_csv('data/grupo.csv')

grupo.to_json('data/grupo.json')
```
